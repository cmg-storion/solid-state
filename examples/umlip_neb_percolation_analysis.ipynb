```python
import os
from ase.io import read
from sevenn.calculator import SevenNetCalculator

from solid_state.nebtools import neb_percolation_barriers, prepare_vasp_neb_folders
```

```python
!mkdir data
!wget -O "./data/Li10Ge(PS6)2_mp.cif" "https://raw.githubusercontent.com/dembart/intro-to-materials-informatics/refs/heads/main/data/seminar09/Li10Ge(PS6)2_mp.cif"
```

```python
n_images = 7
device = 'cpu'
calculators = [SevenNetCalculator(device=device) for _ in range(n_images)]


atoms = read('./data/Li10Ge(PS6)2_mp.cif')
material_id = 'mp-id-test'
submitdir = f'./data/{material_id}_umlip'

result = neb_percolation_barriers(atoms,
                              mobile_specie='Li',
                              upper_bound=5.0,
                              calculators=calculators,
                              submitdir=submitdir,
                              n_images=n_images,
                              )
result.update({'material_id': material_id})
```

```python
print(result)
```

```python
neb_folders = [f for f in os.listdir(submitdir) if '.neb' in f]
path = f'{material_id}_dft'

for folder in neb_folders:
    trajectory = read(f'{submitdir}/{folder}/trajectory_relaxed.xyz', index = ':')
    prepare_vasp_neb_folders(trajectory, f'./data/{path}/{folder}')
```
